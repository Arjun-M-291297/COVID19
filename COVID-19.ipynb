```python
import pandas as pd
import numpy as np
import folium
data=pd.read_csv("covid_19_india.csv")
data.head(15)
```

Grouped States/UT on each Dates based on total Confirmed cases of each day<br>
Changed format of date to required format and sorted dataframe according to date

```python
gp_date=data.groupby(["Date","State/UnionTerritory"])["Confirmed"].agg("sum")
gp_date=pd.DataFrame(gp_date)
gp_date=gp_date.reset_index()
gp_date["Date"]=pd.to_datetime(gp_date["Date"], format='%d/%m/%y')
gp_date=gp_date.sort_values("Date")
gp_date=gp_date.reset_index(drop=True)
gp_date
```

## CHOROPLETH MAP
Creating subsets of dataframe for each year <br>
Creating a Terrain map of Indian Map using folium <br>
On to which colors are specified based on the Confirm_log attribute on each State/Union Territory<br>
File is then saved appending with the current timestamp

```python
for date in dates:
    temp=gp_date[gp_date['Date']==date]
    t= pd.to_datetime(str(date)) 
    timestring = t.strftime('%d %B, %Y')
    india_map = folium.Map(location=[21.6937, 78.9629],zoom_start=5, tiles='Stamen Toner',height=655)
    india_map.choropleth(
    geo_data=india_geojson,
    threshold_scale=[0, 1, 2, 3,4 ,6],
    data=temp,
    columns=['State/UnionTerritory', 'Confirm_log'],
    key_on='feature.properties.NAME_1',
    fill_color='YlOrRd',
    legend_name=timestring,
    )
    india_map.save('map.html')
    driver.get('file:///F:/pY/Exp/map.html')
    now = datetime.now()
    hms=now.strftime("%H_%M_%S")
    driver.save_screenshot("covidImage.png")
    os.rename('covidImage.png', "covidImage"+hms+".png")
```

```python
path = "F:/pY/Exp/"
image_folder = os.fsencode(path)

filenames = []

for file in os.listdir(image_folder):
    filename = os.fsdecode(file)
    if filename.endswith( ('.png') ):
        filenames.append(filename)

filenames.sort()
images = list(map(lambda filename: imageio.imread(filename), filenames))
imageio.mimsave(os.path.join('covid2_zoom5.gif'), images, duration = 0.6)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for date in dates:
    plt.figure(figsize=(30,28))
    temp=gp_date[gp_date["Date"]==date]
    t= pd.to_datetime(str(date)) 
    timestring = t.strftime('%d %B, %Y')
    temp.sort_values("State/UnionTerritory",inplace=True)
    sns.set(font_scale=2.4)
    g =sns.barplot(x='Confirmed',y='State/UnionTerritory',data=temp)
    for p in g.patches:
        percentage = int(p.get_width())
        x = p.get_x() + p.get_width() + 4
        y = p.get_y() + p.get_height()/1.4
        g.annotate(percentage, (x, y))
    plt.title(timestring)
    now = datetime.now()
    hms=now.strftime("%H_%M_%S")
    plt.savefig("covid"+hms+".png")
```
